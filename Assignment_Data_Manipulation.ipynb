```python
# Importing lib 
import pandas as rData
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# reading Data and creating DataFrame
path = "/content/burritos.csv"
myData = rData.read_csv(path)
myData
```

```python
# basic undertsanding 
myData.info()
```

```python
# selecting some col to conduct the mainpulation 
columnList = ['Location', 'Burrito', 'Cost', 'Volume', 'Tortilla', 'Temp', 'Meat', 'Fillings', 'Meat:filling', 'Uniformity', 'Salsa', 'Synergy', 'Wrap', 'overall']
new_data = myData[columnList]
new_data.head()
```

```python
# barplot before handling missing values
corr = new_data.corr()['overall'].reset_index()
corr = corr.sort_values(by = 'overall')[:(len(corr)-1)]
ax = sns.barplot(y = corr['index'], x = corr['overall'], palette='Greens')
ax.set_xlabel('Correlation')
ax.set_ylabel('Features')
ax = ax.set_title('Correlation between Overall Score and features')
```

 -- Determine the total amount of missing values â–¶

```python
# checking the sum of missing_values in all cols
print(myData.isna().sum().sum())
```

-- Choose a column and replace NaN values with the mean â–¶

```python
# first step: finding mean to check 
mean_value = myData['Yelp'].mean()
print(mean_value)

# filling the nan values with the mean
myData['Yelp'].fillna(value=myData['Yelp'].mean(), inplace=True)
myData
```

-- Choose a column and replace NaN values with the median â–¶

```python
# first step: finding median to check 
median_value = myData['Google'].median()
print(median_value)
# filling the nan values with the median
myData['Google'].fillna(value=myData['Google'].median(), inplace=True)
myData
```

-- Choose a column and replace NaN values with the mode â–¶

```python
# first step: finding mode to check 
mode_value = myData['Cost'].mode()
print(mode_value)

# filling the nan values with the median
myData['Cost'].fillna(value=myData['Cost'].mode(), inplace=True)
myData
```

-- Choose a column and replace NaN values with the string â–¶

```python
# filling the nan values in url with a string 'unavailable' 
myData['URL'].fillna('unavailable', inplace=True)
myData
```

-- Choose a column and replace NaN values with the backward filling â–¶

```python
myData['Meat'].bfill(axis ='rows')
```

```python
myData['Tortilla'].ffill(axis ='rows')
```
